```python
#fuction to calculate entropy using 'unique' method
def entropy(class_label_data):
    values,counts=np.unique(class_label_data,return_counts=True)
    for i in range(len(values)):
        # -Pi log2 Pi
        entropy=np.sum([(-counts[i]/np.sum(counts))*np.log2(counts[i]/np.sum(counts))])
    return entropy
```

```python
#calculating information gain
def infogain(data,split_attribute_name,class_label="quality"):
    total_entropy=entropy(data[class_label])
    vals,counts=np.unique(data[split_attribute_name],return_counts=True)
    #weighted entropy
    for i in range(len(vals)):
        weighted_entropy=np.sum([(counts[i]/np.sum(counts))*entropy(data.where(data[split_attribute_name]==vals[i]).dropna()[class_label])])
    information_gain=total_entropy-weighted_entropy
    return information_gain
```

```python
# implementing decision tree algorithm
def ID3(data,originaldata,features,class_label="quality",parent_node_class=None):
    # if all class_label has same value,return the value
    if len(np.unique(data[class_label]))<=1:
        return np.unique(data[class_label])[0]
    
    # if dataset is empty or below some threshold value,then terminate recursion
    elif len(data)==0:
        return np.unique(originaldata[class_label])[np.argmax(np.unique(originaldata[class_label],return_counts=True)[1])]
    
    # if feature space is empty,then terminate recursion
    elif len(features)==0:
        return parent_node_class
    
    #form subtrees if all conditions mentioned above are false
    else:
        # majority class label
        parent_node_class=np.unique(data[class_label])[np.argmax(np.unique(data[class_label],return_counts=True)[1])]
    # select features which best splits the dataset
    for feature in features:
        item_values=[infogain(data,feature,class_label)]
    best_feature_index=np.argmax(item_values)
    best_feature=features[best_feature_index]
    
   # create tree struct
    tree={best_feature:{}}
  # remove the feature with best information gain
    features=[i for i in features if i!=best_feature]
 
  #form subtrees by calling ID3 function recursively
    for value in np.unique(data[best_feature]):
        value=value
        sub_data=data.where(data[best_feature]==value).dropna()
        subtree=ID3(sub_data,dataset,features,class_label,parent_node_class)
        
        #add subtree
        tree[best_feature][value]=subtree
    return (tree)
    
  
```

```python
def predict(query,tree,default=1):
    for key in list(query.keys()):
        if key in list(tree.keys()):
            try:
                result=tree[key][query[key]]
            except:
                return default
            result=tree[key][query[key]]
            if isinstance(result,dict):
                return predict(query,result)
            else:
                return result
```

```python
def test(data,tree):
  # create new query instances by simply removing the target feature column from original dataset & convert into a dictionary
    queries=data.iloc[:,:-1].to_dict(orient="records")
  
  # create an empty dataframe in which prediction of tree are stored
    predicted= pd.DataFrame(columns=["predicted"])
    
  # calculating prediction accuracy
    for i in range(len(data)):
        predicted.loc[i,"predicted"]=predict(queries[i],tree,1.0)
    print("Accuracy of the prediction : ",(np.sum(predicted["predicted"]==data["quality"])/len(data))*100,'%')
    
    
```

```python
from sklearn.model_selection import KFold
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score
```

```python
x=dataset.iloc[:,:-1]
y=dataset.iloc[:,-1]

k=3
kfold=KFold(n_splits=k,random_state=None)
model=LogisticRegression(solver='liblinear')

acc_score=[]

for train_index,test_index in kfold.split(x):
    x_train,x_test=x.iloc[train_index,:],x.iloc[test_index,:]
    y_train,y_test=y[train_index],y[test_index]
    
    model.fit(x_train,y_train)
    pred_values=model.predict(x_test)
    
    acc=accuracy_score(pred_values,y_test)
    acc_score.append(acc)
    
avg_acc_score=sum(acc_score)/k


```

```python
print('Accuracy of each fold : '.format(acc_score))
print('Average accuracy : '.format(avg_acc_score))
```
